```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define the auto-correlation function
def auto_correlation(signal):
  corr=np.correlate(signal,signal,mode='full')
  return corr[len(corr)//2:]  # first of all slice to half the continue 0 to 1/2..

#Define the function for cross-correlation.
def cross_correlation(signal1,signal2):
  corr=np.correlate(signal1,signal2,mode='full')
  return corr[len(corr)//2:]

#Define two signal...
signal1=np.array([1,2,305,4,5])
signal2=np.array([20,4,60,8,1005])

#Compute the correlation...
auto_corr=auto_correlation(signal1)
cross_corr=cross_correlation(signal1,signal2)

print(auto_corr)
print(cross_corr)
```

```python
#Plotting the input signals...
fig,(ax1,ax2) =plt.subplots(2,1,figsize=(10,6))
ax1.plot(signal1)


ax2.plot(signal2)

plt.show()
```

```python
# plot the correlated signals.
fig,(ax1,ax2) =plt.subplots(2,1,figsize=(10,6))
ax1.plot(auto_corr)
ax1.set_title('Auto Correlated Signal...')

ax2.plot(cross_corr)
ax2.set_title('Cross Correlated Signal...')

plt.show()
```
