```python
import numpy as np
import matplotlib.pyplot as plt

n=np.arange(-10,11)

#unit Sample Sequence
unit_sample=np.zeros_like(n)
unit_sample[n==0]=1

#Unit Step Sequence
unit_step=np.heaviside(n,0)

#unit ramp Sequence
unit_ramp=np.maximum(n,1)

#Plotting the signals.
fig,axs=plt.subplots(3,1,figsize=(8,8))

axs[0].stem(n,unit_sample)
axs[0].set_title('Unit Sample Sequence.')

axs[1].stem(n,unit_step) #del u
axs[1].set_title('Unit Step Sequence')

axs[2].stem(n,unit_ramp)
axs[2].set_title('Unit Ramp Sequence')

plt.tight_layout()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate some example data
x = np.linspace(0, 2*np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)

# Create a figure with three subplots arranged vertically
fig, axs = plt.subplots(3, 1, figsize=(16, 16))

# Plot data in the first subplot
axs[0].plot(x, y1, color='red')
axs[0].set_title('Sin(x)')
axs[0].set_xlabel('X-axis')
axs[0].set_ylabel('Y-axis')

# Plot data in the second subplot
axs[1].plot(x, y2, color='blue')
axs[1].set_title('Cos(x)')
axs[1].set_xlabel('X-axis')
axs[1].set_ylabel('Y-axis')

# Plot data in the third subplot
axs[2].plot(x, y3, color='green')
axs[2].set_title('Tan(x)')
axs[2].set_xlabel('X-axis')
axs[2].set_ylabel('Y-axis')

# Adjust spacing between subplots to avoid overlap
plt.tight_layout()

# Display the figure with subplots
plt.show()
```
