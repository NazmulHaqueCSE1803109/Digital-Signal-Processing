```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style

```

**Convolution with 768 x 768 points**

```python
Srate = 256 #Hz
t = np.arange(0,3,1/Srate)
pnts =len(t)
Srate2 = 120 #Hz
t2 = np.arange(0,3,1/Srate)
pnts2 =len(t2)

```

```python
x=[]
h=[]
mat=[]
for i in range(0,len(t)):
  x.append(t[i])

print(x)

for i in range(0,len(t2)):
  h.append(t2[i])

print(h)

for i in range(0,len(h)):
  a=[]
  for j in range(0,len(x)):
    a.append((int)(h[i]*x[j]))

  mat.append(a)


y=[]

for i in range(0,len(h)):
  sum=mat[i][0]
  j=(int)(i)
  k=(int)(0)
  while j-1>=0 and k+1<len(x):
    j=j-1
    k=k+1
    sum=sum+mat[j][k]
  y.append(sum)

for i in range(1,len(x)):
  sum=mat[len(h)-1][i]
  j=(int)(len(h)-1)
  k=(int)(i)
  while j-1>=0 and k+1<len(x):
    j=j-1
    k=k+1
    sum=sum+mat[j][k]
  y.append(sum)
```

```python
#plot the correlated signals
fig, (ax1, ax2) =plt.subplots(2,1, figsize=(10,6))
ax1.plot(t)
ax1.set_title('Signal 1')

ax2.plot(t2)
ax2.set_title('Signal 2')
plt.tight_layout()
plt.show()
```

```python
#plot the correlated signals
fig, ax1 =plt.subplots(1,1, figsize=(10,6))
ax1.plot(y)
ax1.set_title('Signal After convolution')

plt.show()
```
