```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
```

# Applications of convolutions

```python
#Create Signal....
Srate=256 # Hz #Srate typically represents the sampling rate, which is the number of samples obtained per second when digitizing an analog signal.
t=np.arange(0,3,1/Srate)
pnts=len(t)
print(pnts )
print(t)
```

```python
#Create a noiseless Signal
x=np.sin(2*np.pi*2*t) #sin2pi2t same as sin theta

# Create a randomo Image
noise =5*np.random.randn(pnts)

# Adding noise to orginal signal
NoisySignal=x+noise  # it is linear signal because noise is add with x....
# separation of linear noise is very easier....
# if signal is nonlinear then separation is very easier.. environmental signal geneeally non-linear noise.
```

```python
# Ploting orginal signal
plt.figure(figsize=(20,12))
style.use('dark_background')

plt.plot(t,x,label='Orginal Signal')
plt.legend(fontsize=30)

plt.show()
```

```python
# Ploting noisy signal
plt.figure(figsize=(20,12))
style.use('dark_background')

plt.plot(t,NoisySignal,label='Noisy Signal')
plt.legend(fontsize=30)

plt.show()
```

```python
# Ploting noisy signal
plt.figure(figsize=(20,12))
style.use('dark_background')

plt.stem(t,NoisySignal,label='Noisy Signal') #stem is used for discritization...
# A stem plot is a type of plot where each data point is represented by a vertical line (stem) originating from the x-axis at the position of the data point.
plt.legend(fontsize=30)

plt.show()
```

```python
# Selecting filters/
filter=2*np.ones(50)/10.0
print(filter)
```

```python

print(len(filter))
```

```python
# Selecting filters/
FilterSignal=np.convolve(NoisySignal,filter,mode='full') # convolution...
print(filter)
```

```python
# Ploting noisy signal
plt.figure(figsize=(20,12))
style.use('dark_background')

plt.plot(FilterSignal,label='Filtered Signal') #stem is used for discritization...
plt.legend(fontsize=30)

plt.show()
```

```python
#Double filter
FilterSignal2=np.convolve(FilterSignal,filter,mode='full') # convolution...
print(filter)
```

```python
# Ploting noisy signal
plt.figure(figsize=(20,12))
style.use('dark_background')

plt.plot(FilterSignal2,label='Filtered Signal') #stem is used for discritization...
plt.legend(fontsize=30)

plt.show()
```

```python
#Tripple filter
FilterSignal3=np.convolve(FilterSignal2,filter,mode='full') # convolution...
print(filter)
```

```python
# Ploting noisy signal
plt.figure(figsize=(20,12))
style.use('dark_background')

plt.plot(FilterSignal3,label='Filtered Signal') #stem is used for discritization...
plt.legend(fontsize=30)

plt.show()
```

```python
# Ploting noisy signal
plt.figure(figsize=(20,12))
style.use('dark_background')

plt.stem(FilterSignal3,label='Filtered Signal') #stem is used for discritization...
plt.legend(fontsize=30)

plt.show()
```
