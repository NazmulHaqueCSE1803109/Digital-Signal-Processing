```python
import numpy as np
import matplotlib.pyplot as plt

def dft(x):
    N = len(x)
    X = np.zeros(N, dtype=complex)

    for k in range(N):
        for n in range(N):
            X[k] += x[n] * np.exp(-2j * np.pi * k * n / N)

    return X

# Input sequence
x = np.array([1, 1, 0, 0])

# Compute DFT
X = dft(x)


# Compute magnitude and phase spectra
magnitude = np.abs(X)
phase = np.angle(X)

# Plotting
plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.stem(np.arange(len(x)), x, use_line_collection=True)
plt.xlabel('n')
plt.ylabel('x(n)')
plt.title('Input Sequence')

plt.subplot(1, 2, 2)
plt.stem(np.arange(len(magnitude)), magnitude, use_line_collection=True)
plt.xlabel('k')
plt.ylabel('|X(k)|')
plt.title('Magnitude Spectrum')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def dtft(x, a, omega):
    N = len(x)
    X = np.zeros(len(omega), dtype = complex)

    for k in range(len(omega)):
        for n in range(N):
            X[k] += (x[n] * np.exp(-1j * omega[k] * n))

    return X


# Input Sequence parameters
a = 0.9
n = np.arange(-10, 11)
x = a ** n

# DTFT parameters
omega = np.linspace(-np.pi, np.pi, 1000)

#Compute DTFT
X = dtft(x, a, omega)

# Compute magnitude and phase spectre
magnitude = np.abs(X)
phase  = np.angle(X)

# Plotting
plt.figure(figsize = (10, 4))

plt.subplot(1, 2, 1)
plt.stem(n, x)
plt.xlabel('n')
plt.ylabel('x(n)')
plt.title('Input Sequence')

plt.subplot(1, 2, 2)
plt.stem(omega, magnitude)
plt.xlabel('k')
plt.ylabel('|X(k)|')
plt.title('Magnitude Spectrum')

plt.tight_layout()
plt.show()
```
