```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define the auto-correlation function
def auto_correlation(signal):
  corr=np.correlate(signal,signal,mode='full')
  return corr[len(corr)//2:]  # first of all slice to half the continue 0 to 1/2..

#Define the function for cross-correlation.
def cross_correlation(signal1,signal2):
  corr=np.correlate(signal1,signal2,mode='full')
  return corr[len(corr)//2:]

#Define two signal...

# Parameters for the first sinusoidal signal
f1 = 1  # Frequency of the first signal (in Hz)
A1 = 1  # Amplitude of the first signal
phase1 = 0  # Phase of the first signal (in radians)

# Parameters for the second sinusoidal signal
f2 = 2  # Frequency of the second signal (in Hz)
A2 = 0.5  # Amplitude of the second signal
phase2 = np.pi/2  # Phase of the second signal (in radians)

# Time range
t = np.linspace(0, 1, 1000)  # Time values from 0 to 1 second with 1000 samples

# Generate the first sinusoidal signal
signal1 = A1 * np.sin(2 * np.pi * f1 * t + phase1)

# Generate the second sinusoidal signal
signal2 = A2 * np.sin(2 * np.pi * f2 * t + phase2)

#Compute the correlation...
auto_corr=auto_correlation(signal1)
cross_corr=cross_correlation(signal1,signal2)

print(auto_corr)
print(cross_corr)
```

```python
#Plotting the input signals...
fig,(ax1,ax2) =plt.subplots(2,1,figsize=(10,6))
ax1.plot(signal1)


ax2.plot(signal2)

plt.show()
```

```python
# plot the correlated signals.
fig,(ax1,ax2) =plt.subplots(2,1,figsize=(10,6))
ax1.plot(auto_corr)
ax1.set_title('Auto Correlated Signal...')

ax2.plot(cross_corr)
ax2.set_title('Cross Correlated Signal...')

plt.show()
```
