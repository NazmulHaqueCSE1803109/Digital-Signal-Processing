# Correlation between two signals

```python
Srate = 256 #Hz
t = np.arange(0,4,1/Srate)
pnts =len(t)
Srate2 = 120 #Hz
t2 = np.arange(0,4,1/Srate)
pnts2 =len(t2)
```

```python

mat=[]


for i in range(0,len(t)):
  a=[]
  for j in range(0,len(t2)):
    a.append((int)(t[i]*t2[j]))

  mat.append(a)

y=[]


for i in range(len(t)-1,-1,-1):
  sum=mat[i][0]
  j=(int)(i)
  k=(int)(0)
  while j+1<len(t) and k+1<len(t2):
    j=j+1
    k=k+1
    sum=sum+mat[j][k]
  y.append(sum)

for i in range(1,len(t2)):
  sum=mat[0][i]
  j=(int)(0)
  k=(int)(i)
  while j+1<len(t) and k+1<len(t2):
    j=j+1
    k=k+1
    sum=sum+mat[j][k]
  y.append(sum)

print(y)
```

```python
#plot the correlated signals
fig, (ax1, ax2) =plt.subplots(2,1, figsize=(20,10))
ax1.plot(t)
ax1.set_title('Signal 1')

ax2.plot(t2)
ax2.set_title('Signal 2')

plt.show()
```

```python
#plot the correlated signals
fig, ax1 =plt.subplots(1,1, figsize=(20,10))
ax1.plot(y)
ax1.set_title('Signal After correlation')

plt.show()
```

# Z-Transformation

```python
def Z_Transformation(signal,n):
  N=len(signal)
  Z=np.zeros(N,dtype=complex)

  for k in range(N):
    Z[k]=np.sum(signal**(-n)) #np.exp(-1j*2*np.pi*n*k/N)

  roc ='Entire Z-plane'
  if np.abs(Z).any():
    roc='ROC: |z| <{}'.format(np.max(np.abs(Z)))
  return Z,roc

n = np.arange(0, 1024)
signal = 0.5**n * (n >= 0)


Z,roc=Z_Transformation(t2,n)

#Plotting ROC.
fig ,ax =plt.subplots()
circle=plt.Circle((0,0),radius=np.max(np.abs(Z)),edgecolor='r')
ax.add_patch(circle)
plt.xlim(-np.max(np.abs(Z))-1,np.max(np.abs(Z))+1)
plt.ylim(-np.max(np.abs(Z))-1,np.max(np.abs(Z))+1)

plt.xlabel('Real')
plt.ylabel('Imaginary')
plt.title("Region of Convergence")
plt.grid(True,linestyle='--',linewidth=0.5)
plt.show()

#Printing the values of Z.
print('Z-Transformation values...')
for i,z in enumerate(Z):
  print('Z[{}]:{}'.format(i,z))
print(roc)
```
